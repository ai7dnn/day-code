```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()
```

```python
# Load an example dataset
tips = sns.load_dataset("tips")
tips
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
);
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker");
```

```python
sns.lmplot(data=penguins, x="bill_length_mm", y="bill_depth_mm");
```

```python
sns.lmplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue='island');
```

```python
sns.lmplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue='species');
```

```python
sns.lmplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    hue="species", col="sex", height=3,
);
```

```python
sns.lmplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    col="species", row="sex", height=3,
);
```

```python
sns.lmplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    col="species", row="sex", height=3,
    facet_kws=dict(sharex=False, sharey=False),
);
```
